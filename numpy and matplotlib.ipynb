```python

image = plt.imread('img2.jpg')
gray_image = np.dot(image[..., :3], [0.2989, 0.5870, 0.1140])

bw_image = np.where(gray_image > 127, 255, 0)
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(bw_image, cmap='gray')
plt.title('BW Image')
plt.axis('off')

plt.show()
```

```python


print('Input image shape:', image.shape) # always do it 

if len(image.shape) == 2:
    #  grayscale, -> RGB
    image = np.stack((image,) * 3, axis=-1)
elif image.shape[2] != 3:
    # RGB
    image = image[:, :, :3]

border_size = 1000

padded_image = np.pad(image, pad_width=((border_size, border_size), (border_size, border_size), (0, 0)), mode='constant', constant_values=0)

plt.imshow(padded_image)
plt.axis('off')
plt.show()
```

```python
al = 2
blended_image = al * image +(1-al) * image
plt.imshow(blended_image)
plt.axis('off')
plt.show()
```

```python
rescaled_image = np.kron(image, np.ones((2, 2)))
plt.imshow(rescaled_image)
plt.axis('on')
plt.show()
```
